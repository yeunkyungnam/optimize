```python
from ortools.linear_solver import pywraplp
import openpyxl
import os
import sys
import time
import datetime
from datetime import datetime as dt

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker
import numpy as np
import pandas as pd
```

```python
solver = pywraplp.Solver.CreateSolver('GLOP')
```

```python
DEM = pd.read_excel('data.xlsx',sheet_name='DEM')
CAPA = pd.read_excel('data.xlsx',sheet_name='CAPA')
PCOST = pd.read_excel('data.xlsx',sheet_name='PCOST')
SCOST = pd.read_excel('data.xlsx',sheet_name='SCOST')
ISTOCK = pd.read_excel('data.xlsx',sheet_name='ISTOCK')
```

```python
ITEM = DEM['ITEM'].drop_duplicates().to_list()
RES = CAPA['RES'].drop_duplicates().to_list()
TIME = DEM['TIME'].drop_duplicates().to_list()
```

```python
prod = {}
store = {}
for i in range(len(ITEM)):
    prod[i]=[[solver.NumVar(0, solver.infinity(), 'prodqty[%i][%i][%i]' %(i,j,k)) for k in range(len(TIME))] for j in range(len(RES))]

for i in range(len(ITEM)):
    store[i] = [[solver.NumVar(0, solver.infinity(), 'storeqty[%i][%i]' %(i,k)) for k in range(len(TIME))]]

        
print('Number of variables =', solver.NumVariables())
```
